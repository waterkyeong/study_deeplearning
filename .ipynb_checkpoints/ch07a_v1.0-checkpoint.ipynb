```python
train = df[:-4]
test = df[-4:]
```

```python
historical_mean = np.mean(train["data"])
historical_mean
```

```python
test.loc[:, "pred_mean"] = historical_mean
```

```python
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mape_historical_mean = mape(test["data"], test["pred_mean"])
mape_historical_mean
```

```python
fig, ax = plt.subplots()

ax.plot(train["date"], train["data"], "g-", label="Train")
ax.plot(test["date"], test["data"], "b-", label="Test")
ax.plot(test["date"], test["pred_mean"], "r--", label="Predicated")

ax.set_xlabel("Date")
ax.set_ylabel("per share (USD)")
ax.axvspan(80,83, color="#808080", alpha=0.2)
ax.legend(loc=2)

plt.xticks(
    np.arange(0, 85, 8),
    [1960, 1962, 1964, 1966, 1968, 1970, 1972, 1974, 1976, 1978, 1980]
)

fig.autofmt_xdate()
plt.tight_layout
```

## 참고사항

```python
def simulate_process(is_stationary):
    np.random.seed(42)
    process = np.empty(400)
    
    if is_stationary:
        alpha = 0.5
        process[0] = 0
    else:
        alpha = 1
        process[0] = 10

    for i in range(400):
        if i + 1 < 400:
            process[i + 1] = alpha * process[i] + np.random.standard_normal()
        else:
            break

    return process
```

```python
stationary = simulate_process(True)
non_stationary = simulate_process(False)
```

```python
fig, ax = plt.subplots()

ax.plot(stationary, linestyle="-", label="stationary")
ax.plot(non_stationary, linestyle="--", label="stationary")
ax.set_xlabel("Timesteps")
ax.set_ylabel("Value")
plt.tight_layout
```

```python
np.random.seed(42)
steps = np.random.standard_normal(1000)
steps[0] = 0
random_walk = np.cumsum(steps)
```

```python
fig, ax = plt.subplots()

ax.plot(random_walk, linestyle="-", label="stationary")
ax.set_xlabel("Timesteps")
ax.set_ylabel("Value")
plt.tight_layout
```

```python
ADF_result = adfuller(random_walk)
print(ADF_result[0], ADF_result[1])
```

```python
plot_acf(random_walk, lags=20)
```

```python
diff_random_walk = np.diff(random_walk, n=1)
```

```python
fig, ax = plt.subplots()

ax.plot(diff_random_walk, linestyle="-", label="stationary")
ax.set_xlabel("Timesteps")
ax.set_ylabel("Value")
plt.tight_layout
```

```python
ADF_result = adfuller(diff_random_walk)
print(ADF_result[0], ADF_result[1])
```

```python
plot_acf(diff_random_walk, lags=20)
```

## 교재 7.2(예제)
- ARIMA 오류로 인해서 SARIMAX로 교체

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

from datetime import datetime
```

```python
series = pd.read_csv("data/sales.csv", index_col=0, parse_dates=[0], date_format="%Y-%m")
series.head()
```

```python
X = series.values
X = np.nan_to_num(X)
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = []
```

```python
for t in range(len(test)):
    model = SARIMAX(history, order=(5,1,0))
    model_fit = model.fit(disp=False)
    output = model_fit.predict()
    yhat = output.mean()
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print(f"{yhat}, {obs}")
```

```python
error = mean_squared_error(test, predictions)
print(error)
```

```python
plt.plot(test)
plt.plot(predictions)
```
